## Operaciones Básicas

```python
from sklearn.datasets import load_breast_cancer

# Cargar el conjunto de datos
data = load_breast_cancer()

# Convertir los datos en un DataFrame de pandas
cancer_df = pd.DataFrame(data.data, columns=data.feature_names)

# Agregar la columna de destino (cáncer maligno o benigno)
cancer_df['breast cancer type'] = data.target_names[data.target]
```

## Combinaciones

- PAM50.csv contiente una lista de genes y proteinas usada por el sistema de clasificación PAM50. La columna RefSeqProteinID contiene el ID de la proteina que puede ser trazado junto con otros datasets.

- clinical.csv contiene datos clinicos de cancer de seno asociados a un proteoma.

- proteomes.csv contiene información sobre el nivel de expresión de diferentes proteínas en el proteoma analizado, en escala logarítmica.

Para más contexto sobre el origen de estos datos se puede dirigir a:
https://www.kaggle.com/datasets/piotrgrabo/breastcancerproteomes


```python
"""
Siga las siguientes instrucciones
A medidas que avance en cada paso, observe como cambia el resultado
"""

# 1. Lea el archivo "proteomes.csv" en forma de un DataFrame llamado "proteomes"
proteomes_df = pd.read_csv('proteomes.csv', decimal=',')

# 2. Declare la columna "RefSeq_accession_number" como el índice, y guardelo (pista; inplace)
proteomes_df.set_index('RefSeq_accession_number',  inplace = True)

# 3. Use transpose (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transpose.html) para transponer proteomes
proteomes_df_transposed = proteomes_df.transpose()

# Muestre las primeras 5 filas de proteomes
proteomes_df_transposed.head(5)
```

```python
"""
Siga las siguientes instrucciones
A medidas que avance en cada paso, observe como cambia el resultado
"""

# 1. Lea el archivo "clinical.csv" en forma de un DataFrame llamado "clinical"
clinical_df = pd.read_csv('clinical.csv', decimal=',')

# 2. Declare la columna "Complete TCGA ID" como el índice, y guardelo (pista; inplace)
clinical_df.set_index('Complete TCGA ID',  inplace = True)

# Muestre las primeras 5 filas de clinical
clinical_df_transposed = clinical_df.transpose()

clinical_df.head(5)
```

\Realice la combinación de clinical y proteomes empleando join, concat y merge

```python
# Unir dataframes a través de un Index
proteomes_df_transposed.join(clinical_df)
```

```python
# Concatenar los DataFrames a lo largo del eje de las columnas
pd.concat([clinical_df, proteomes_df_transposed], axis=1)
```

```python
pd.merge(clinical_df, proteomes_df_transposed, left_index=True, right_index=True)
```
